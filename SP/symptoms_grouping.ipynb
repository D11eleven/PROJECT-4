```python
# Dependencies
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

# Attempting to Group Symptoms

```python
# Name of the CSV file
file = '../Resources/training_dataset_original.csv'
```

```python
# The correct encoding must be used to read the CSV in pandas
data_df = pd.read_csv(file, encoding="ISO-8859-1", low_memory=False)
```

```python
data_df.head()
```

```python
symptom_counts = data_df.groupby(data_df['SYMPTOM']).count().sort_values('VAERS_ID', ascending=False)
symptom_counts
```

```python
symptom_counts.to_csv('../Resources/symptom_counts.csv', index=True)
```

# Merging in Symptom Groups and Severity

```python
# Name of the CSV file
groupingfile = '../Resources/symptom_grouping_worksheet.csv'
```

```python
group_df = pd.read_csv(groupingfile)
```

```python
group_df
```

```python
group_df.dropna(axis=0, how = 'any', inplace = True)
group_df
```

```python
organized_df = data_df.merge(group_df)
organized_df.head()
```

```python
clean_df = organized_df[['VAERS_ID', 'SYMPTOM', 'ASSIGNED_GROUP', 'SEVERITY_LEVEL', 'HOSPITAL','AGE_YRS', 'OTHER_MEDS', 'CUR_ILL',
                        'HISTORY', 'PRIOR_VAX', 'ALLERGIES', 'F', 'M', 'JANSSEN', 'MODERNA', 'PFIZER',
                        'VAX_DOSE_SERIES_1', 'VAX_DOSE_SERIES_2', 'VAX_SITE_LA', 'VAX_SITE_RA']]
clean_df
```

```python
clean_df.to_csv('../Resources/FINAL_CLEAN_FILE.csv', index=False)
```
